## 1.1 Read the intersected complexes files

```python
def read_all_complexes_from_region(loop_id, p2complexes_root, position = 'Left'):
#     p2complexes_root = '../../results/intersection_files/CTCF_ALL/'

#     middle_region_folder = 'region_intersection_complex'
#     side_region_folder = '{}_100_region_intersection_complex'.format(position.lower())
#     p2complexes_root = '../../results/intersection_files/'

    p2middle_region_complexes = os.path.join(p2complexes_root, 
#                                              middle_region_folder, 
                                             '{}_middle_100_region.csv'.format(loop_id))
    p2side_region_complexes = os.path.join(p2complexes_root, 
#                                           side_region_folder,
                                          '{}_{}_100_region.csv'.format(loop_id, position.lower()))

    df_complex_middle = pd.read_csv(p2middle_region_complexes, 
                                    index_col=0).drop_duplicates()
    df_complex_side = pd.read_csv(p2side_region_complexes, 
                                    index_col=0).drop_duplicates()
    
    if position == 'Left':
        mid_idx_overlap_motif = (df_complex_middle['complex_direction'] == 'Left')
        side_idx_overlap_motif = (df_complex_side['complex_direction'] == 'Right')       
    else:
        mid_idx_overlap_motif = (df_complex_middle['complex_direction'] == 'Right')      
        side_idx_overlap_motif = (df_complex_side['complex_direction'] == 'Left')
        
    df_complex_middle = df_complex_middle[mid_idx_overlap_motif]    
    df_complex_side = df_complex_side[side_idx_overlap_motif]
    
    return df_complex_middle, df_complex_side
```

# 2. Convert intersected fragment into array and align gems in both side

```python
def convert_frag_str_to_array(List_of_fragment):
    # split by fragments
    list_of_frag_coord = List_of_fragment.split(';')
    
    # get rid of chr<>, split number into a tuple.
    list_of_frag_coord = [x.split(':')[1] for x in list_of_frag_coord]
    list_of_frag_coord = [x.split('-') for x in list_of_frag_coord]
    
    # order the fragment by starting site.
    list_of_frag_coord.sort(key = lambda x: int(x[0]))
    
    # convert fragment to integer. 
    # Note that end site has '(x)' at the end.
    list_of_frag_coord = [[int(x[0]), int(x[1].split('(')[0])] for x in list_of_frag_coord]
    frag_coord_filtered_array = np.array(list_of_frag_coord)
    
    return frag_coord_filtered_array
```

```python
def align_mid_and_side_to_mid(df_mid, df_side, drop_side = True):
    '''
    Here df_mid and df_side come from same region. Only needs to align the GEM_ID
    depends on drop_side, return either:
        True: select from df_side only aligned complexes, extended frag_list in df_mid
        False: concat the un-aligned complexes in df_side to extended df_mid
        
    NOTE that the joint frag list does not need to be in order. as convert_str_to_array did it.
    '''
    side_shared_gem_id_index = df_side['GEM_ID'].isin(df_mid['GEM_ID'])
    list_of_match_gem_id = df_side['GEM_ID'][side_shared_gem_id_index]
    
    for gem_id in list_of_match_gem_id:
        side_gem_index = df_side['GEM_ID'] == gem_id
        match_complex_frag_list = df_side[side_gem_index]['Intersected_list_of_frag_coordinates'].values[0]
        
        mid_gem_index = df_mid['GEM_ID'] == gem_id
        mid_complex_frag_list = df_mid[mid_gem_index]['Intersected_list_of_frag_coordinates'].values[0]
        
        joint_frag_list = ';'.join([match_complex_frag_list, mid_complex_frag_list])
        df_mid.loc[mid_gem_index, 'Intersected_list_of_frag_coordinates'] = joint_frag_list
    
    # remove the matched complexes from df_side.
    df_side = df_side[~side_shared_gem_id_index]
    if drop_side:
        return df_mid
    else:
        return pd.concat((df_mid, df_side))
```

# 3. Plot functions

```python
def plot_parallel_lines(list_of_gems, ax = None, left_most_site = 0, right_most_site = 100, 
                        left_motif_region_in = (0, 5), 
                        right_motif_region_in = (95, 100), 
                        motif_extended = False,
                        sorted_by_len = False,
                        title = 'ordered GEM', silent = False,
                       color = 'b'):
    '''
    plot a list of gems in parallel,
    where each row is a gem: 
        [[frag1_start, frag1_end], [frag2_start, frag2_end], ..., [fragk_start, fragk_end]]
    so the list if 
    1. plot a straight line. thin, from start to end.
    2. plot each frag. thick, from its start to end.
    3. plot vertically left most and right most site.
    '''
    if not sorted_by_len:
        if not isinstance(list_of_gems, list):
            sorted_list_of_gems = list_of_gems.tolist()
        sorted_list_of_gems.sort(key = lambda x: x[-1][-1] - x[0][0])
    else:
        if not isinstance(list_of_gems, list):
            sorted_list_of_gems = list_of_gems.tolist()
    
    y_idx = np.arange(len(list_of_gems))[::-1]
    
    y_buffer = 1
    x_buffer = (right_most_site - left_most_site) * 0.05
    
    total_length = right_most_site - left_most_site
    if not motif_extended:
        extension = 4000 * np.array([-1, 1])
    else:
        extension = np.array([0, 0])
        
    left_motif_region = left_motif_region_in + extension
    right_motif_region = right_motif_region_in + extension
 
    left_motif_percent = (left_motif_region[1] - left_motif_region[0])/total_length * 100
    right_motif_percent = (right_motif_region[1] - right_motif_region[0])/total_length * 100
    
#     height = (len(y_idx) + y_buffer * 2) * 0.1
# #     minimum_height_to_avoid_cutting_title = 2
# #     if height <= minimum_height_to_avoid_cutting_title:
# #         height = minimum_height_to_avoid_cutting_title
# #         y_buffer = (height / 0.1 - len(y_idx)) // 2
#     width = 6
    
#     if ax is None:
#         fig, ax = plt.subplots(1, figsize = (width, height), dpi = 150)
#     else:
#         ax.figure.set_size_inches(width, height)
    
    for idx, line in enumerate(sorted_list_of_gems):
        
        left_end = line[0][0]
        right_end = line[-1][-1]
        x_thin = [left_end, right_end]
        y_thin = np.ones_like(x_thin) * y_idx[idx]
        ax.plot(x_thin, y_thin, c = 'grey', alpha = 0.8, lw = 0.8)
        for frag in line:
            x_thick = frag
            if (x_thick[-1] - x_thick[0]) < (0.002 * total_length):
                x_thick[-1] = x_thick[0] + 0.002 * total_length 
            y_thick = np.ones_like(x_thick) * y_idx[idx]
            
#             colorcode #0000B2 for CTCF ChIA-Drop and #005900 for cohesin ChIA-Drop.
            ax.plot(x_thick, y_thick, c = color, lw = 3.8)
    
#     # 3. add vertical lines.
#     ax.axvspan(left_motif_region[0], left_motif_region[1], alpha = 0.3, color = 'r')
#     ax.axvspan(right_motif_region[0], right_motif_region[1], alpha = 0.3, color = 'r')
    
    # set limits on x & y axis
    ax.set_ylim( y_idx.min() - y_buffer, y_idx.max() + y_buffer)    
    ax.set_xlim(left_most_site - x_buffer, right_most_site + x_buffer)
    
    # change x & y axis ticks and labels.
#     xlabels = ['left motif_ext\n({:.2f}%)'.format(left_motif_percent),
#               'right motif_ext\n({:.2f}%)'.format(right_motif_percent)]
    xlabels = ['left motif'.format(left_motif_percent),
              'right motif'.format(right_motif_percent)]
    ax.set_xticks(ticks = [np.mean(left_motif_region),
                        np.mean(right_motif_region)])
    ax.set_xticklabels(labels = xlabels, fontsize = 6)
    ax.tick_params(axis = 'y', which = 'both', left = False, top = False, 
                   labelleft = False)
    
#     ax.set_title(title, fontsize = 8)
    
    if not silent:
        plt.show()
#     plt.savefig('tmp.png')
    return ax
        
    
```

```python
def plot_mid_and_side_separatly(df_loop_annot, loop_id, p2complexes, position = 'Left', 
                      sep = True, drop_side = True, title = 'ordered_gem', savefig = False,
                               color = '#0000B2'):
    
    df_mid, df_side = read_all_complexes_from_region(loop_id, p2complexes, position)
    
    idx_loop = df_loop_annot['loop_ID'] == loop_id
    
    left_most_site, right_most_site = df_loop_annot[idx_loop][['left_start', 
                                                                'right_end']].values[0]
    loop_length = right_most_site - left_most_site
    
    left_motif_region = df_loop_annot[idx_loop][['left_motif_start', 
                                                 'left_motif_end']].values[0]
    right_motif_region = df_loop_annot[idx_loop][['right_motif_start', 
                                                  'right_motif_end']].values[0]
    
    if position == 'Left':
        left_most_site -= loop_length
    elif position == 'Right':
        right_most_site += loop_length
    
    if sep:
        df_combine = pd.concat((df_mid, df_side))
        sorted_by_len = False
    else:
        df_combine = align_mid_and_side_to_mid(df_mid.copy(), df_side.copy(),
                                              drop_side = drop_side)
        sorted_by_len = False
    mid_num, comb_num, side_num = df_mid.shape[0], df_combine.shape[0], df_side.shape[0]
    share_num = mid_num + side_num - comb_num
    mid_only_num = mid_num - share_num
    side_only_num = side_num - share_num
    num_of_complex_list = [side_only_num, share_num, mid_only_num, 0, 0]
    if position == 'Right':
        num_of_complex_list = num_of_complex_list[::-1]
        
    gem_count = 'left_{}_leftMiddle_{}_middle_{}_middleRight_{}_right_{}'.format(*num_of_complex_list)
    
    title = '{}_{}\n_{}'.format(title, loop_id, gem_count)
    
    mid_only_idx = ~df_combine.GEM_ID.isin(df_side.GEM_ID)
    side_only_idx = ~df_combine.GEM_ID.isin(df_mid.GEM_ID)
    
    df_mid_only = df_combine[mid_only_idx]
    df_side_only = df_combine[side_only_idx]
    df_across_boarder = df_combine[~(mid_only_idx | side_only_idx)]
    
    list_of_plotting_df = [df_mid_only, df_side_only, df_across_boarder]

    fig_name_list = [f'{loop_id}_{x}_ordered_gem' for x in [f'mid_{position.lower()}', 
                                                            position.lower(), 
                                                            f'across_mid_{position.lower()}']]
    list_of_list_of_gems = [cur_df['Intersected_list_of_frag_coordinates'].apply(convert_frag_str_to_array) for \
                           cur_df in list_of_plotting_df]
    Heights = [len(list_x) for list_x in list_of_list_of_gems]
    
    width, height = (6, sum(Heights) * 0.1)
    fig, axes = plt.subplots(3, 1, dpi = 150, figsize = (width, height), 
                             gridspec_kw={'height_ratios': Heights, 
                                          'hspace': 0},
                            sharex = True)
    
    fig.subplots_adjust(top = 0.9 * (height + 0.2)/height)
    for idx, cur_df in enumerate(list_of_plotting_df):        
#         list_of_gems = cur_df['Intersected_list_of_frag_coordinates'].apply(convert_frag_str_to_array)
        list_of_gems = list_of_list_of_gems[idx]
        ax = axes[idx]
        if len(list_of_gems) == 0:
            ax.tick_params(axis = 'y', which = 'both', left = False, top = False, 
                   labelleft = False)
            continue
            
#         fig, ax = plt.subplots(1, dpi = 120)
        
        ax = plot_parallel_lines(list_of_gems, ax = ax, 
                                 left_most_site = left_most_site, right_most_site = right_most_site, 
                                 left_motif_region_in = left_motif_region, 
                                 right_motif_region_in = right_motif_region, 
                                 motif_extended = False,
                                 sorted_by_len = sorted_by_len,
                                 title = '', 
                                 silent = True, 
                                color = color)
#     fig.subplots_adjust(vspace = 0)
#     axes[0].set_title(title)
    
    fig.suptitle(title, fontsize = 8, y = 0.9 * (height + 0.6)/height)
#     fig.subplots_adjust(top = 0.9)
    
    if savefig:
#         plt.tight_layout(rect=[0, 0.05, 1, 0.95])
#         plt.tight_layout(rect=[0, 0.05, 1, 0.9 * (height + 0.2)/height])
        plt.savefig('../../../../new_data_0422_2020/new_results_0526/sample_loop_ordered_gem_plot/CTCF_0.2_PASS_all_filtered_loops_height_adjusted/' + \
                    f'{loop_id}_mid_{position.lower()}_combined_ordered_gem.png', 
                    dpi = 150,
                   bbox_inches='tight')
        plt.close()
    else:
#         plt.tight_layout(rect=[0, 0.05, 1, 0.9 * (height + 0.2)/height])
        plt.tight_layout(rect=[0, 0.05, 1, 0.95])
        plt.show()
        plt.close()

    return None
```

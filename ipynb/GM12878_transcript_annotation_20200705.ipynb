```python
def read_gtffile(directory, file_name):
    with open(directory + file_name) as f:
        gtf = {}
        for line in f:
            tmp = line.strip().split("\t")
            if '#' not in tmp[0]:
                if tmp[2] == 'transcript':
                    if tmp[0]=='MT':
                        tmp[0] = 'M'
                    chrom = 'chr'+tmp[0]
                    start = int(tmp[3])
                    end = int(tmp[4])
                    strand = tmp[6]
                    geneid = tmp[8].split(";")[0].split('"')[1]
                    transid = tmp[8].split(";")[2].split('"')[1]
                    genename = tmp[8].split(";")[4].split('"')[1]
                    gtf[transid] = [chrom, start, end, strand, transid, geneid, genename]
    return gtf
```

```python
def read_gtffile2(directory, file_name, gtf):
    with open(directory + file_name) as f:
        #gtf = {}
        for line in f:
            tmp = line.strip().split("\t")
            if tmp[2] == 'transcript':
                chrom = tmp[0]
                start = int(tmp[3])
                end = int(tmp[4])
                strand = tmp[6]
                geneid = tmp[8].split(";")[0].split('"')[1]
                transid = tmp[8].split(";")[1].split('"')[1]
                genename = tmp[8].split(";")[2].split('"')[1]
                if transid not in gtf.keys():
                    gtf[transid] = [chrom, start, end, strand, transid, geneid, genename]
    return gtf
```

```python
def read_transfile(directory, file_name):
    with open(directory + file_name) as f:
        transcript = {}
        next(f)
        for line in f:
            tmp = line.strip().split("\t")
            transid = tmp[0].split(".")[0]
            if 'ENST' in transid:
                transcript[transid] = [float(tmp[5]), float(tmp[6])]
    return transcript
```

```python
def write_result(directory, out_list, out_name):
    with open(directory+out_name, 'a') as file1:
        for i in range(len(out_list)):
            file1.write('\t'.join(map(str, out_list[i])) + '\n')
    file1.close()
```
